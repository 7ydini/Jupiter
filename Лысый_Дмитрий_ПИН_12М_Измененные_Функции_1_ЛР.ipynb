```python
time = [
    1, 2, 3, 4,
    5, 6, 7
    ]
```

```python
def nf(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    result[i] = n * time[i]
  return result
```

```python
def nn(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    result[i] = n * time[i] * n * time[i]
  return result
```

```python
def nsqrt(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    try:
      result[i] = math.sqrt(n * time[i])
    except OverflowError:
      result[i] = result[i-1] * 2
  return result
```

```python
def nnn(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    result[i] = math.pow(n * time[i], 3)
  return result
```

```python
def npow2(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    try:
      # result[i] = math.log2(n * time[i])
      result[i] = math.pow(2, n * time[i])
    except OverflowError:
      result[i] = 0
  return result
```

```python
def lgn(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    try:
      result[i] = math.log(n * time[i], 10)
    except OverflowError:
      result[i] = 0
  return result
```

```python
def nlgn(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    try:
      result[i] = math.log(n * time[i], 10) * n
    except OverflowError:
      result[i] = 0
  return result
```

```python
def nfact(n):
  result = [0, 0, 0, 0, 0, 0, 0]
  for i in range(len(time)):
    try:
      result[i] = math.factorial(time[i]*n)
    except OverflowError:
      result[i] = 0
  return result
```

```python
n = 1
times = ["Second", "Minute", "Hour", "Day", "Month", "Year", "Century"]
fn = ["lg(n)", "n^(1/2)", "n", "n*lg(n)", "n*n", "n^3", "2^n", "n!"]
plt.figure(figsize = (24, 24))

# Plot bars and create text labels for the table

plt.subplot(2, 1, 1)
plt.grid(True, which='both')
plt.minorticks_on()


table_text = [lgn(n),
              nsqrt(n),
              nf(n),
              nlgn(n),
              nn(n),
              nnn(n),
              npow2(n),
              nfact(n)]


colors = plt.cm.BuPu(np.linspace(0, 0.5, len(fn)))
index = np.arange(len(times)) + 0.3
bar_width = 0.4

# colors = colors[::-1]
y_offset = np.zeros(len(times))
for row in range(len(fn)):
    plt.bar(index, table_text[row], bar_width, color=colors[row])
    y_offset = y_offset + table_text[row]
table = plt.table(cellText=table_text,
                      rowLabels=fn,
                      colLabels=times,
                      )
# Adjust layout to make room for the table:
plt.subplots_adjust(left=0.2, bottom=0.2)
table.set_fontsize(50)
table.scale(1, 3)
plt.xticks([])
```
